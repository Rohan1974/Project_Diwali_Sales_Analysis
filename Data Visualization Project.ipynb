# Exploratory Data Analysis (EDA)

## 1) Gender

```python
df.columns
```

**plotting count plot of gender**

```python
x= sns.countplot(x='Gender', data=df)
```

**counting total number of females and males on the top of bar**

```python
x= sns.countplot(x='Gender', data=df)

for bars in x.containers:
    x.bar_label(bars)
```

## A) Amount Basis

**use group by**

```python
sales_gender=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gender
```

```python
y= sns.barplot(x='Gender', y='Amount', data=sales_gender)

for bars in y.containers:
    y.bar_label(bars)
```

**Conclusion No: 1**
    
**From the above graph we can see most of the buyers are Females and even the purchasing power of females are more than males**

# 2) Age

```python
df.columns
```

```python
x= sns.countplot(x='Age Group',data=df,  hue='Gender')

for bars in x.containers:
    x.bar_label(bars)
```

## A) Amount Basis

```python
sns.barplot(x='Age Group', y='Amount', data=sales_age)

for bars in y.containers:
    y.bar_label(bars)
```

# 7) Top Selling Products By Product_ID and Product_Category

## 1) Product_ID

```python
Top_Prod= df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

X= sns.barplot(y='Product_Category',x='Orders', data=Top_Prod)

for bars in X.containers:
    X.bar_label(bars)
```
